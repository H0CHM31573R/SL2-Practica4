```python
import numpy as np
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-X))

class RedNeuronal:
    def __init__(self, Wl1, Wl2, bl1 = 0, bl2 = 0):
        self.Wl1 = np.array(Wl1)
        self.bl1 = np.array(bl1)

        self.Wl2  = np.array(Wl2)
        self.bl2 = np.array(bl2)

    def forward_propagation(self,X):
        print(X.shape)
        print(self.Wl1.T.shape)
        A = self.bl1 + sigmoid(np.matmul(X,self.Wl1.T))
        print(A)
        print(A.shape)
        y_aprox = sigmoid(self.bl2 + np.matmul(A,self.Wl2))
        print(y_aprox.shape)
        return y_aprox
```

```python
w1 = np.array([
                [5,-2],
                [7, -3]
              ])

b1 = np.array([-8, 1])

w2 = np.array([5, 3])

b2 = -6

X = np.array([
      [1,1],
      [1,0],
      [0,1],
      [0,0],
])
```

```python
RedNeuronal(w1, w2, b1, b2).forward_propagation(X)
```

Se puede observar que, en terminos de magnitud, la primera neurona se activa con mayor intensidad. Sin embargo, esto solo es relevante porque la ultima capa usa una funcion sigmoide como activacion, si se usara algo como ReLU esta neurona serviria potencialmente tendria impacto nulo (o bien, serviria para prededecir un cero) cuando tenga alta magnitud y sentido negativo.

```python
w1 = np.array([
                [0.8, 0.2],
                [0.4, 0.9],
                [0.3, 0.5]
              ])

w2 = np.array([0.3, 0.5, 0.9])

X = np.array([
      [1,1],
      [1,0],
      [0,1],
      [0,0],
])
```

```python
RedNeuronal(w1, w2).forward_propagation(X)
```

En este caso podemos observar una especie de uniformidad en las neuronas, esto quiere decir que, como basicamente todas producen informacion similar, existe cierta redundancia entre ellas. Puede ser que tengamos un overfitting (demasiadas neuronas) o bien que no hayan pasado suficientes epocas de entrenamiento para que las neuronas se puedan diferenciar mas

```python
w1 = np.array([
                [20,20],
                [-20,-20]
              ])

b1 = np.array([-10, 30])

w2 = np.array([20,20])

b2 = -30

X = np.array([
      [1,1],
      [1,0],
      [0,1],
      [0,0],
])
```

```python
RedNeuronal(w1, w2, b1, b2).forward_propagation(X)
```

Aqui vemos una bifuracion bastante clara de las neuronas, una proveyendo gran valor y la otra restando relativamente poco, sin embargo, el comportamiento de las neuronas varia muy poco conforme varian las entradas, en otras palabras tenemos una red neuronal con un gran sesgo
